```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


with open("/kaggle/input/bible-text/asv.txt") as f:
    bible = f.read()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=75,
    length_function=len,
    is_separator_regex=False,
)
texts = text_splitter.create_documents([bible])
    
```

```python
print(bible[0:1000])
```

```python
texts_spl=texts[15:40]
[print(text.page_content+"\n\n") for text in texts_spl]
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings_model=HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
embeddings = embeddings_model.embed_documents(
    [
       text.page_content for text in texts_spl
    ]
)
len(embeddings), len(embeddings[0])
```

```python
embedded_query = embeddings_model.embed_query("What did God do with the rib taken from man")
embedded_query[:5]
```

```python
from langchain_chroma import Chroma

db = Chroma.from_documents(texts_spl, HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
import transformers
import torch
from transformers import AutoModelForCausalLM,AutoTokenizer

# model path
model_id = "/kaggle/input/llama-3/transformers/8b-chat-hf/1"

tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
pipeline = transformers.pipeline(
                                 "text-generation",
                                  model=model_id,
                                  torch_dtype=torch.float16,
                                  device_map="auto")

def response(Question,length):
    sequences = pipeline(
               Question,
               do_sample=True,        # to sample from the models output distribution
               top_k=10,              # to restrict sampling to the top-k most likely next words
               num_return_sequences=1,
               eos_token_id=tokenizer.eos_token_id,
               max_length=length,     # length the generated sequence
               )

    #print('Question: ' , Question + '\n')
    
    if len(sequences)>0:
        print(sequences[0]['generated_text'][len(Question):] + '\n')
    else:
        print("Failed to generate an answer.")
```

```python
query = "What did God do with the rib taken from man"
docs = db.similarity_search(query)
print(docs[0].page_content+"\n\n\n")
print(docs[1].page_content)
```

<h1>So here the question we ask will be regarding man's missing rib and it should answer regarding the retrieved text</h1>

```python
query = input()
start_time = time.time()
docs = db.similarity_search(query)
Question = f"""Answer the question in brief according to the context given:\n
Context:\n\t{docs[0].page_content}\n\t{docs[1].page_content}\n\n\nQuestion:{query}"""
length= 170
answer = response(Question,length)
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Time taken: {elapsed_time} seconds\n")
print(answer)
```
